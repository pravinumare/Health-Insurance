#### Detecting Outliers

```python
# Finding out outliers using zscore method
```

```python
def get_outliers(column):
    
    outliers_list = []
    mean = column.mean()
    std = column.std()
    threshold = 3
    
    for i in column.values:
        zscore = (i - mean)/std
        
        if np.abs(zscore) > threshold:
            outliers_list.append(i)
            
    return outliers_list
```

```python
get_outliers(df['Age(yrs)'])
```

```python
sns.boxplot(x = df['Age(yrs)'])       # Boxplot Method to find out Outliers
```

```python
# iqr (Interquantile Method) method
```

```python
q1 = df['Age(yrs)'].quantile(0.25)
q2 = df['Age(yrs)'].quantile(0.50)
q3 = df['Age(yrs)'].quantile(0.75)

iqr = q3 - q1

upper_tail = q3 + 1.5*iqr
lower_tail = q1 - 1.5*iqr

print("Upper tail :",upper_tail)
print("Lower tail :",lower_tail)
```

```python
# We dont see any outliers here as the minimum and maximum values (0.25 and 85 respective) are within the upper and lower tail
```

### Splitting Data

```python
x = df.drop('Premium', axis = 1)
y = df['Premium']
```

# Model Training

### 1. Linear Regression

```python
linear_model = LinearRegression()
linear_model.fit(x_train, y_train)
```

```python
# Testing Accuracy

y_pred = linear_model.predict(x_test)

mse = mean_squared_error(y_test,y_pred)
print("MSE value", mse)

mae = mean_absolute_error(y_test,y_pred)
print("MAE value", mae)

lin_reg_test_r_value = r2_score(y_test,y_pred)
print("R_squared value", lin_reg_test_r_value)
```

```python
# Training Accuracy

y_train_pred = linear_model.predict(x_train)

mse = mean_squared_error(y_train,y_train_pred)
print("MSE value", mse)

mae = mean_absolute_error(y_train,y_train_pred)
print("MAE value", mae)

lin_reg_train_r_value_ = r2_score(y_train,y_train_pred)
print("R_squared value", lin_reg_train_r_value_)
```

### 2. K Nearest Neighbour

```python
# We need to scale the data for KNN Regressor
```

```python
norm_scaler = MinMaxScaler()
norm_array = norm_scaler.fit_transform(x_train)
x_train_norm_df = pd.DataFrame(norm_array, columns=x.columns)
# x_train_norm_df
```

```python
norm_array = norm_scaler.transform(x_test)
x_test_norm_df = pd.DataFrame(norm_array, columns=x.columns)
# x_test_norm_df
```

```python
knn_model = KNeighborsRegressor()
knn_model.fit(x_train_norm_df, y_train)
```

```python
# Testing accuracy

y_pred = knn_model.predict(x_test_norm_df)

mse = mean_squared_error(y_test,y_pred)
print("MSE value", mse)

mae = mean_absolute_error(y_test,y_pred)
print("MAE value", mae)

knn_test_r_value = r2_score(y_test,y_pred)
print("R_squared value", knn_test_r_value)
```

```python
# Training accuracy

y_pred = knn_model.predict(x_train_norm_df)

mse = mean_squared_error(y_train,y_pred)
print("MSE value", mse)

mae = mean_absolute_error(y_train,y_pred)
print("MAE value", mae)

knn_train_r_value = r2_score(y_train,y_pred)
print("R_squared value", knn_train_r_value)
```

### Hyperparameter Tuning

```python
knn_reg = KNeighborsRegressor()

hyperparameters = {'n_neighbors':np.arange(2,31), 'p':[1,2]}

rscv_knn_model = RandomizedSearchCV(knn_reg, hyperparameters, cv = 5)
rscv_knn_model.fit(x_train_norm_df, y_train)
```

```python
rscv_knn_model.best_estimator_
```

```python
knn_model_hyp = rscv_knn_model.best_estimator_
knn_model_hyp.fit(x_train_norm_df, y_train)
```

```python
# Testing accuracy

y_pred = knn_model_hyp.predict(x_test_norm_df)

mse = mean_squared_error(y_test,y_pred)
print("MSE value", mse)

mae = mean_absolute_error(y_test,y_pred)
print("MAE value", mae)

knn_hyp_test_r_value = r2_score(y_test,y_pred)
print("R_squared value", knn_hyp_test_r_value)
```

```python
# Training accuracy

y_pred = knn_model_hyp.predict(x_train_norm_df)

mse = mean_squared_error(y_train,y_pred)
print("MSE value", mse)

mae = mean_absolute_error(y_train,y_pred)
print("MAE value", mae)

knn_hyp_train_r_value = r2_score(y_train,y_pred)
print("R_squared value", knn_hyp_train_r_value)
```

```python
def age(birthday):
    today = date.today()
    DOB = datetime.strptime(birthday, '%d/%m/%Y')

    age = today.year - DOB.year
    return age
```

```python
Age = age(Birthdate)
Age
```

```python
def gender_encoding(gender):
    if gender == 'male':
        gender = 0
        
    elif gender == 'female':
        gender = 1
        
    return gender
```

```python
gender = gender_encoding(Gender)
gender
```

```python
def BMI(height, weight):
    bmi = weight / ((height/100)**2)
    return bmi
```
